Machine Learning Model

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
```

```python
# Combine ingredients and cuisine into a single string
df['Combined_Features'] = df['Cleaned-Ingredients'].apply(lambda x: ' '.join(x) if isinstance(x, list) else str(x)) +' ' + df['Cuisine']
```

```python
# Create a TF-IDF vectorizer for text features
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['Combined_Features'])
```

```python
# Scale numerical features
scaler = MinMaxScaler()
df['Ingredient-count_scaled'] = scaler.fit_transform(df[['Ingredient-count']])
df['TotalTimeInMins_scaled'] = scaler.fit_transform(df[['TotalTimeInMins']])
df['Servings_scaled'] = scaler.fit_transform(df[['Servings']])
df['Remarks_scaled'] = scaler.fit_transform(df[['Remarks']])
```

```python
def recommend_dishes(ingredients, cuisine, ingredient_count, total_time,servings, remarks, num_recommendations=5):
  """Recommends dishes based on input features."""

  # Transform ingredients and cuisine into a vector
  features = ' '.join(ingredients) + ' ' + cuisine
  features_vector = tfidf.transform([features])

  # Calculate cosine similarity for text features
  cosine_similarities = cosine_similarity(features_vector, tfidf_matrix)

  # Calculate weighted similarity based on numerical features
  ingredient_count_diff = np.abs(df['Ingredient-count_scaled'].values - ingredient_count)
  total_time_diff = np.abs(df['TotalTimeInMins_scaled'].values - total_time)
  servings_diff = np.abs(df['Servings_scaled'].values - servings)
  remarks_diff = np.abs(df['Remarks_scaled'].values - remarks)
  weighted_similarity = cosine_similarities * (1 - ingredient_count_diff) * (1 - total_time_diff) * (1 - servings_diff) * (1 - remarks_diff)

  # Get indices of top recommendations
  top_indices = np.argsort(weighted_similarity[0])[::-1][:num_recommendations]

  # Create a DataFrame for recommendations
  recommendations = df.iloc[top_indices][['Name', 'Cuisine']]
  recommendations['Similarity'] = weighted_similarity[0][top_indices]

  return recommendations
```

```python
# Split data into training and test sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
def evaluate_model(df, num_recommendations=5):
  """Evaluates the recommendation model."""

  correct_recommendations = 0
  total_dishes = len(df)

  for index, row in df.iterrows():
    ingredients = row['Cleaned-Ingredients']
    cuisine = row['Cuisine']
    ingredient_count = row['Ingredient-count_scaled']
    total_time = row['TotalTimeInMins_scaled']
    servings = row['Servings_scaled']
    remarks = row['Remarks_scaled']
    actual_dish = row['Name']

    recommendations = recommend_dishes(ingredients, cuisine, ingredient_count, total_time,servings,remarks, num_recommendations)
    recommended_dishes = recommendations['Name'].tolist()

    if actual_dish in recommended_dishes:
      correct_recommendations += 1

  accuracy = correct_recommendations / total_dishes
  return accuracy
```

```python
# Evaluate the model
accuracy = evaluate_model(test_df)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# Get user input
ingredients = input("Enter ingredients separated by commas: ").split(',')
ingredients = [x.strip().lower() for x in ingredients]
cuisine = input("Enter cuisine: ").strip().lower()
ingredient_count = float(input("Enter ingredient count: "))
total_time = float(input("Enter total time in minutes: "))
servings = int(input("Enter number of servings: "))
remarks = int(input("Enter remarks code: ")) # Assuming remarks is an integer code

# Scale user input
ingredient_count_scaled = scaler.transform([[ingredient_count]])[0][0]
total_time_scaled = scaler.transform([[total_time]])[0][0]
servings_scaled = scaler.transform([[servings]])[0][0]
remarks_scaled = scaler.transform([[remarks]])[0][0]

# Recommend dishes
recommendations = recommend_dishes(ingredients, cuisine, ingredient_count_scaled, total_time_scaled,servings_scaled,remarks_scaled)

# Print recommendations
print(recommendations)
```
